```python
import pandas as pd
```

```python
df = pd.read_csv("response.csv", sep=",")
```

```python
header_template = "Basandoti sulle informazioni presenti nel documento fornito dai un punteggio alle seguenti risposte relative alla domanda: \"{question}\"\n\n"
body = "Risposta {letter}:\n{response}\n\n"
```

```python
evaluation_query = ""

for label, content in df.iterrows():
    for id, item in enumerate(content.items()):
        if id == 0:
            evaluation_query += header_template.format(question=item[1])
        else:
            evaluation_query += body.format(letter=id, response=item[1])
    evaluation_query += "\n\n\n***************************************\n\n\n"
print(evaluation_query)
```
